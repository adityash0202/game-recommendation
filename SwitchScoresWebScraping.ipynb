```python
import requests
import string
from bs4 import BeautifulSoup

website_url=[]

for letter in string.ascii_uppercase:
    website_url.append('https://www.switchscores.com/games/by-title/'+letter)

print("Websites need to be Scraped:")
for each_website in website_url:
  print(each_website)
```

```python
#!/bin/bash

# Import the required libraries
import requests
from bs4 import BeautifulSoup


# Create empty lists to store the scraped data
title = []
release_date = []
price = []
rating = []

# Loop through each website URL
for each_website in website_url:
  
  # Get the HTML content of the website and parse it using BeautifulSoup
  website = requests.get(each_website).text
  soup = BeautifulSoup(website,'lxml')
  
  # Extract the game data and rating using BeautifulSoup
  games_data = soup.find_all('div', class_='col-xs-9 col-sm-9 col-md-9 col-lg-9')
  games_rating = soup.find_all('div', class_='col-xs-3 col-sm-3 col-md-3 col-lg-3')
  
  # Print the scraped data to check if it's working correctly
  print(games_data)
  print(games_rating)
  
  # Extract the game title, release date, and price
  for each_game in games_data:
    title.append(each_game.find('span', class_='h5').text.strip())
    release_date.append(each_game.find('span', class_='h6').text.split("Released:")[1].strip())
    if each_game.find('span', class_='h4') is not None:
      price.append(each_game.find('span', class_='h4').text.strip())
    else:
      price.append('Nan')
    
  # Extract the game rating
  for each_game in games_rating:
    try:
      rating.append(float(each_game.span.text.strip()))
    except ValueError: 
      fraction=each_game.span.text.strip()
      numerator, denominator = fraction.split('/')
      result = float(numerator) / float(denominator)
      rating.append(round(result*10,2))
  
```

```python
import pandas as pd
game_df = pd.DataFrame({'Title': title, 
                           'Release': release_date,
                           'Price': price,
                          'Rating': rating
                        })
game_df
     
```
