Read files to merge

```python
wikipedia_data = pd.read_csv('wikipedia.csv')
switch_scores_data = pd.read_csv('SwitchScores.csv')
```

```python
wikipedia_data.head()
```

Removing wikipedia_data rows with plots having nan values

```python
wikipedia_data.Plots.isnull().sum()
```

```python
wikipedia_data = wikipedia_data.dropna(subset=['Plots'])
```

```python
wikipedia_data.Plots.isnull().sum()
```

Switch Scores website Data

```python
switch_scores_data.head()
```

Merging different websites web scraped data 

```python
merged_data = pd.merge(wikipedia_data, switch_scores_data,on='Title',how='left')
merged_data
```

Removing columns which are not useful

```python
merged_data = merged_data.drop('Unnamed: 0_y', axis=1)
merged_data = merged_data.drop('Release Date', axis=1)
merged_data = merged_data.drop('Unnamed: 0_x', axis=1)
merged_data
```

Cleaning Price Data

```python
# Convert the 'Price' column to a float data type after removing the '£' symbol
merged_data['Price'] = merged_data['Price'].str.replace('£', '').astype(float)
```

```python
# Get the indices of rows with NaN values in the 'Price' column
nan_prices_idx = merged_data.index[merged_data['Price'].isna()]

# Fill NaN values in 'Price' column with mean price for games from the same publisher
for idx in nan_prices_idx:
    publisher = merged_data.loc[idx, 'Publisher']
    mean_price = merged_data[merged_data['Publisher'] == publisher]['Price'].mean()
    merged_data.loc[idx, 'Price'] = mean_price
```

```python
# Replace 0.00 values in 'price' columns with NaN
merged_data['Price'].replace(0.00, np.nan, inplace=True)

# Fill NaN values in 'price' columns with column means
merged_data['Price'].fillna(merged_data['Price'].mean(), inplace=True)

merged_data
```

Cleaning Ratings data

```python
# Get the indices of rows with NaN values in the 'Rating' column
nan_ratings_idx = merged_data.index[merged_data['Rating'].isna()]

# Fill NaN values in 'Rating' column with mean rating for games from the same publisher
for idx in nan_ratings_idx:
    publisher = merged_data.loc[idx, 'Publisher']
    mean_rating = merged_data[merged_data['Publisher'] == publisher]['Rating'].mean()
    merged_data.loc[idx, 'Rating'] = mean_rating
```

```python
# Replace 0.00 values in 'rating' columns with NaN

merged_data['Rating'].replace(0.00, np.nan, inplace=True)

# Fill NaN values in 'rating' columns with column means

merged_data['Rating'].fillna(merged_data['Rating'].mean(), inplace=True)

merged_data
```
